```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

def collatz(n, plot=True):
    array = []
    array.append(n)
    while n != 1:
        if n%2 == 0:
            n= n/2
        else:
            n = 3*n+1
        array.append(n)

    array = np.array(array)
    if plot:
        fig = plt.plot(array, lw=3)
        plt.scatter(np.arange(0,len(array)), array, color="black")
        plt.grid(alpha=0.2, linestyle="--", color="gray")
        plt.xlabel("Iteration number")
        plt.ylabel("Value")
        plt.title("The infinite collatz conjecture")
        return array, max(array), fig
    else:
        return array, max(array)
```

```python
collatz(100, plot=True)
```
